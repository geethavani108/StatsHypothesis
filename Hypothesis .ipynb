```python
from scipy.stats import shapiro
#data_to_test=data['volatile acidity']
data_to_test=data['fixed acidity']
stat, p = shapiro(data_to_test)
print("stat =%.2f, p =%.30f" %(stat,p))
if p < 0.05 :
    print (" Not a Normal Distribution")
else :
    print("Normal Distribution")
```

# Non Parametric Test  -- Mann-Whitney U test (wilcoxon ) - Tests whether the distribution of two or more independent samples are equal or not

# Assumptions -- Identical distribution, observations can be ranked

```python
TreatmentA=[3,4,2,6,2,5]
TreatmentB=[9,7,5,10,6,8]
Ustats,p =scipy.stats.mannwhitneyu(TreatmentA,TreatmentB)
print("stat =%.3f, p =%.3f" %(Ustats,p))
if p > 0.05 :
    print ("Same Distribution")
else :
    print("Different Distribution")
    
```
